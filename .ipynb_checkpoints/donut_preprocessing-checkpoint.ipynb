```python
import json
import os
import shutil
import random
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
# ====== CONFIG ======
LABEL_STUDIO_JSON = "./annotations/manish_annotations.json"  # Label Studio export
IMAGES_SRC_FOLDER = "./images/manish_images"
OUTPUT_FOLDER = "./dataset/donut/train"

SPLIT_RATIOS = (0.8, 0.1, 0.1)  # train/val/test split
FONT_PATH = "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf"  # Update if needed
```

```python
# Define schema fields
GLOBAL_FIELDS = [
    "Supplier Name", "Receipt No.", "Receipt Date", "Currency", "Transaction ID", "Payment Method",
    "Special Instructions", "VAT Code", "VAT Percent", "VAT Amount",
    "Sub Total", "Net Amount", "Total Discount", "Coupon Name", "Coupon Amount",
    "Total Amount", "Change Amount", "Total Item Count", "Sale Amount",
    "Payment Date", "Payment Mode", "Card Details"
]

ITEM_FIELDS = [
    "Item Code", "Item Name", "Item Quantity", "Item Unit Price", "Item Discount",
    "Item Amount", "Item VAT Code"
]
```

```python
# ====== CREATE OUTPUT STRUCTURE ======
os.makedirs(os.path.join(OUTPUT_FOLDER, "images"), exist_ok=True)
os.makedirs(os.path.join(OUTPUT_FOLDER, "labels"), exist_ok=True)
```

```python
def visualize_image(image_path, results):
    img = Image.open(image_path).convert("RGB")
    draw = ImageDraw.Draw(img)
    font = ImageFont.truetype(FONT_PATH, 16)

    for res in results:
        if res["type"] in ["rectanglelabels", "polygonlabels"]:
            points = res["value"].get("points", [])
            if len(points) >= 2:
                x1 = points[0]["x"] * img.width / 100
                y1 = points[0]["y"] * img.height / 100
                x2 = points[1]["x"] * img.width / 100
                y2 = points[1]["y"] * img.height / 100
                label = res["value"].get("rectanglelabels", res["value"].get("polygonlabels", [""]))[0]
                draw.rectangle([x1, y1, x2, y2], outline="red", width=2)
                draw.text((x1, y1 - 15), label, font=font, fill="blue")

    plt.figure(figsize=(10, 12))
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
# ====== LOAD LABEL STUDIO JSON ======
with open(LABEL_STUDIO_JSON, "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
def normalize_key(label):
    return label.strip().lower().replace(" ", "_")
```

```python
def extract_text(res):
    # Safely get text value from Label Studio
    if "text" in res["value"] and res["value"]["text"]:
        return res["value"]["text"][0]
    elif "transcription" in res["value"] and res["value"]["transcription"]:
        return res["value"]["transcription"][0]
    return ""
```

```python
converted = []

for item in data:
    image_path = item["data"]["image"].replace("file://", "")
    image_name = os.path.basename(image_path)
    label_file = os.path.splitext(image_name)[0] + ".json"

    gt_parse = {normalize_key(f): "" for f in GLOBAL_FIELDS}
    gt_parse["items"] = []

    results = item["annotations"][0].get("result", []) if "annotations" in item else []

    global_fields = {}
    item_candidates = []

    for res in results:
        if res["type"] in ["rectanglelabels", "polygonlabels"]:
            value = res["value"]
            # Extract label name
            label = value.get("rectanglelabels", value.get("polygonlabels", []))[0]
            label_clean = normalize_key(label)

            # Extract associated text
            transcription = extract_text(res)

            # Extract position
            points = value.get("points", [])
            y = points[0]["y"] if points else 0
            x = points[0]["x"] if points else 0

            # Debug check
            print(f"Found label: {label_clean}, text: {transcription}")

            if label in GLOBAL_FIELDS:
                global_fields[normalize_key(label)] = transcription
            elif label in ITEM_FIELDS:
                item_candidates.append({"label": normalize_key(label), "text": transcription, "y": y, "x": x})

    # Fill global fields
    for k, v in global_fields.items():
        gt_parse[k] = v

    # Group item fields by row (y-coordinate)
    item_candidates.sort(key=lambda x: (round(x["y"], 1), x["x"]))
    grouped_items = {}
    for item_data in item_candidates:
        row_key = round(item_data["y"], 1)
        if row_key not in grouped_items:
            grouped_items[row_key] = {}
        grouped_items[row_key][item_data["label"]] = item_data["text"]

    for _, fields in sorted(grouped_items.items()):
        gt_parse["items"].append(fields)

    converted.append({
        "image": image_path,
        "image_name": image_name,
        "label_file": label_file,
        "gt_parse": gt_parse,
        "results": results
    })
```

```python
# ====== VISUALIZE SAMPLE ======
print("Showing one sample for validation...")
visualize_image(converted[10]["image"], converted[0]["results"])
print("Grouped items for this image:", json.dumps(converted[0]["gt_parse"]["items"], indent=2))

# ====== DATASET SPLIT ======
random.shuffle(converted)
n = len(converted)
train_end = int(n * SPLIT_RATIOS[0])
val_end = train_end + int(n * SPLIT_RATIOS[1])

splits = {
    "train": converted[:train_end],
    "val": converted[train_end:val_end],
    "test": converted[val_end:]
}

# Create folders and save files
for split, items in splits.items():
    img_dir = os.path.join(OUTPUT_FOLDER, split, "images")
    lbl_dir = os.path.join(OUTPUT_FOLDER, split, "labels")
    os.makedirs(img_dir, exist_ok=True)
    os.makedirs(lbl_dir, exist_ok=True)

    for entry in items:
        shutil.copy(entry["image"], os.path.join(img_dir, entry["image_name"]))
        with open(os.path.join(lbl_dir, entry["label_file"]), "w", encoding="utf-8") as f:
            json.dump({"gt_parse": entry["gt_parse"]}, f, ensure_ascii=False, indent=2)

print(f"✅ Dataset split complete: {len(splits['train'])} train, {len(splits['val'])} val, {len(splits['test'])} test.")
```
