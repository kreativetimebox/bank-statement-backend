```python
# !pip install transformers==4.36.2 torch torchvision Pillow

import os
from transformers import DonutProcessor, VisionEncoderDecoderModel
from PIL import Image
import json

# Load Donut model and processor
processor = DonutProcessor.from_pretrained("naver-clova-ix/donut-base-finetuned-docvqa")
model = VisionEncoderDecoderModel.from_pretrained("naver-clova-ix/donut-base-finetuned-docvqa")

def extract_receipt_data(image_file):
    image = Image.open(image_file).convert("RGB")
    prompt = "Extract key information from receipt."
    inputs = processor(image, text=prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=256, num_beams=4, early_stopping=True)
    result = processor.decode(outputs[0], skip_special_tokens=True)
    try:
        json_output = json.loads(result)
    except Exception:
        json_output = {"raw_text": result}
    return json_output

# List all your receipt image file paths explicitly here
image_files = [
    "/kaggle/input/recepit/Receipt_Image/10042.png",
    # Add all other image file paths here
]

all_results = []
for img_path in image_files:
    print(f"Processing: {img_path}")
    data = extract_receipt_data(img_path)
    print(json.dumps(data, indent=2))
    all_results.append({"filename": os.path.basename(img_path), "data": data})

# Save results to JSON file
with open("all_receipt_results.json", "w") as f:
    json.dump(all_results, f, indent=2)

print(f"Extraction finished for {len(all_results)} receipts. Results saved to all_receipt_results.json.")
```
